## Ejercicio 3: Reducción de dimensionalidad.

Utilizando la matriz obtenida en el ejercicio anterior:
1. Aplique `PCA` para obtener $n$ componentes principales de la matriz, donde `n = min(20, X.shape[0])`. ¿Es necesario estandarizar o escalar los datos?
2. Grafique la varianza capturada por los primeros $n$ componentes principales, para cada $n$.
3. En base al gráfico, seleccione las primeras $m$ columnas de la matriz transformada para agregar como nuevas características al conjunto de datos.

#### 3.1 Rta//: Construimos un conjunto de datos que solo cuente con las variables numéricas continuas, para aplicar el PCA. A su vez dicho conjunto cuenta con un escalado sobre la matriz original para asegurar de que todas las variables estén en las mismas unidades y ninguna tenga un peso demasiado grande.

#### - Por tanto excluimos las variables: 
  - Date.
  - Suburb.
  - Type.
  - Postcode.

#### Nota: si bien Postcode se presenta como variable númerica, dado que esta representa una categoría que discrimina un territorio fue excluída.

````python
# Sin categoricas
melb_df2 = pd.read_csv('https://github.com/Grupo242022/EyCD/blob/main/melb_data_extended.csv?raw=true')
filtered_df=melb_df2[melb_df2.columns.difference(['Date','Suburb','Type', 'Postcode','CouncilArea'])]
feature_dict_sc = list(filtered_df.T.to_dict().values())
vec_sc = DictVectorizer(sparse=False) 
feature_matrix_sc = vec_sc.fit_transform(feature_dict_sc)

# df_enc_sc - Encoding sin variables categóricas:
df_enc_sc = pd.DataFrame(feature_matrix_sc, columns=vec_sc.get_feature_names())
df_enc_sc.head(1)
````

````python
from sklearn import preprocessing 

scaler_sc = preprocessing.MinMaxScaler()

# Data frame escalado sin categoricas
df_esc_sc = scaler_sc.fit_transform(df_enc_sc)  #aplicando
esc_sc_df = pd.DataFrame(df_esc_sc, columns=vec_sc.get_feature_names())

imp_sc = IterativeImputer(random_state=0, estimator=KNeighborsRegressor(),n_nearest_features=5)
values_sc = esc_sc_df.values.copy()

# Ajuste del metodo de imputer a la Matriz escalada - min max:
scaled_knn_sc = imputer.fit_transform(values_sc)

imp_sc = pd.DataFrame(scaled_knn_sc, columns=vec_sc.get_feature_names())
````

### 3.2 Ahora si aplicamos PCA a los datos númericos escalados y obtenemos la varianza capturada por los primeros $n$ componentes principales, para cada $n$.:

#### * Decidimos que vamos a estandarizar (buscamos que las variables tengan media cero y desviación estándar de uno) antes de realizar el estudio PCA. Así evitaremos que alguna de las variables del conjunto que tenga una media y dispersión muy superior al resto, domine la mayoría de las componentes principales.

````python
# Para este ejercicio tomamos parte del codigo y explicacion de
# https://www.cienciadedatos.net/documentos/py19-pca-python.html 
# ==============================================================================

# Entrenamiento modelo PCA con escalado de los datos
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler


n = min(20, imp_sc.shape[1])

# n_components = determina el número de componentes calculados.
pca_pipe = make_pipeline(StandardScaler(),PCA(n_components=n)) # Estandarizado
pca_pipe.fit(imp_sc)

# Se extrae el modelo entrenado del pipeline
modelo_pca = pca_pipe.named_steps['pca']
````

#### * Ingresando la información de las componentes creadas:

````python
# Construimos el data frame de la transformación por componentes principales PCA
# ==============================================================================

pca_data = pd.DataFrame(
    data    = modelo_pca.components_,
    columns = vec_sc.get_feature_names(),
    index   = ['PC1', 'PC2', 'PC3', 'PC4','PC5', 'PC6', 'PC7', 'PC8','PC9', 'PC10', 'PC11', 'PC12']
)
pca_data
````

#### - Trasformación: una vez tenemos entrenado el modelo, con el método `transform()`, podemos reducir la dimensionalidad de las nuevas observaciones.

````python
# Proyección de las observaciones de entrenamiento
# ==============================================================================
proyecciones = pca_pipe.transform(X=imp_sc)
proyecciones = pd.DataFrame(
    proyecciones,
    columns = ['PC1', 'PC2', 'PC3', 'PC4','PC5', 'PC6', 'PC7', 'PC8','PC9', 'PC10', 'PC11', 'PC12'],
    index   = imp_sc.index
)
print(f'proyecciones.shape: {proyecciones.shape}')
print('--- --- --- --- Muestra de las proyecciones --- --- --- --- --- ')
proyecciones.head()
````

#### * Al analizar a la primera componente esta es el resultado de la siguiente combinación lineal de las variables originales:


$ PC1 = 0.45*Bathroom	+ 0.38*BuildingArea	+ 0.36*Car + 0.31*Distance + 0.08*Landsize - 0.11*Lattitude +	0.21*Longtitude + 0.31*Price	- 0.07*Propertycount	+ 0.50*Rooms + 0.11*YearBuilt	- 0.10*airbnb\_price\_median$

#### Donde del anterior, se puede resaltar que los pesos asignados en la primera componente a las variables: Bathroom, BuildingArea, Car y Distance es aproximadamente igual entre ellos y superiores al asignado de las demás variables. 

											


### 3.3. En base al gráfico, seleccione las primeras $m$ columnas de la matriz transformada para agregar como nuevas características al conjunto de datos.

````python
# Porcentaje de varianza explicada por cada componente
# ==============================================================================
print('----------------------------------------------------')
print('Porcentaje de varianza explicada por cada componente')
print('----------------------------------------------------')
print(modelo_pca.explained_variance_ratio_)
````

#### Posteriormente, teniendo a las componentes principales calculadas en el modelo, accedemos a la varianza explicada por cada una de ellas, así como a la proporción respecto al total y la proporción de varianza acumulada. Haciendo uso de los atributos: `explained_variance_` y `explained_variance_ratio_`...

````python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 5))
ax.bar(
    x      = np.arange(modelo_pca.n_components_) + 1,
    height = modelo_pca.explained_variance_ratio_
)


#### Graficamos las componentes principales:
for x, y in zip(np.arange(len(esc_sc_df.columns)) + 1, modelo_pca.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )

ax.set_xticks(np.arange(modelo_pca.n_components_) + 1)
ax.set_ylim(0, .25)
ax.set_title('Porcentaje de varianza explicada por cada componente')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada');
````

#### - En conclusión, para este ejercicio, la primera componente explica el 22% de la varianza observada en los datos, mientras que la segunda el 16% y la tercera el 12% . A su vez se nota que las tres últimas componentes no superan por separado el 1% de varianza explicada.

#### * Aplicamos el Encode usando el método de OneHotEncoder:
- Paso 1: Sobre la matriz con datos categóricos. A manera de ejemplo mostramos los primeros 3 registros.

````python
# Step 1: encode categorical columns
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

X_cat = encoder.fit_transform(melb_df_cat)
new_columns = []
for col, col_values in zip(melb_df_cat.columns, encoder.categories_):

  for col_value in col_values:
    new_columns.append('{}={}'.format(col, col_value))

print(f"Matrix has shape {X_cat.shape}, \nwith columns: {new_columns[:3]}", end='\n')
````

````python
# Step 2: Append the numerical columns
X = np.hstack([X_cat, melb_df_num.values])
new_columns.extend(melb_df_num)
print("Matrix has shape {}, with columns: {}".format(X_cat.shape, new_columns[-4:]))
````
