```python
import matplotlib.pyplot as plt
import numpy
import pandas as pd

import seaborn
seaborn.set_context('talk')

from sqlalchemy import create_engine, text
```

Definimos el Dataset "df_melb" con los datos de las viviendas de Melbourne, conjunto de datos de la compentencia Kaggle sobre estimación de precios de ventas de propiedades en Melbourne, Australia.

```python
url = 'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/melb_data.csv'
df_melb = pd.read_csv(url)
```

Creamos una base de datos en SQLite utilizando la libreria SQLalchemy del dataset df_melb.

```python
engine = create_engine('sqlite:///melbourne_data.sqlite3', echo=True)
```

```python
df_melb.to_sql('melb_data', con=engine, if_exists="replace")
```

Gerenamos un nuevo Dataset "df_airbnb" con las columnas de interes vistas en clases.

```python
interesting_cols = [
  'description', 'neighborhood_overview',
  'street', 'neighborhood', 'city', 'suburb', 'state', 'zipcode',
  'price', 'weekly_price', 'monthly_price',
  'latitude', 'longitude',
]
df_airbnb = pd.read_csv(
    'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/cleansed_listings_dec18.csv',
    usecols=interesting_cols)
```

Creamos una base de datos en SQLite utilizando la libreria SQLalchemy del dataset df_airbnb.

```python
df_airbnb.to_sql('melb_data_airbnb', con=engine, if_exists="replace")
```

Vamos a ver el contenido de "zipcode" de la tabla melb_data_airbnb para corroborar que coincidan con los valores de "Postcode" de melb_data con la que haremos el JOIN

```python
query4 = "SELECT zipcode, COUNT(1) as Records_of_zipcodes FROM melb_data_airbnb GROUP BY zipcode ORDER BY Records_of_zipcodes DESC "
with engine.connect() as con:
      rs = con.execute(query4)
      print(query4)
      for row in rs:
          print(row)

      print('\n\n')
```

Se observan que los dos resultados con mayor frecuencia son 3000.0 con 2491 apariciones y 3000 con 876, siendo en realidad el mismo código postal.<br>
Además, filtramos por zipcode y calculamos el promedio de los precios diarios, semanales y mensuales.

```python
query5 = """
SELECT zipcode,  
    AVG(price) as price_mean, 
    AVG(weekly_price) as weekly_price_mean, 
    AVG(monthly_price) as monthly_price_mean
FROM (
    SELECT 
        CAST(zipcode AS float) as zipcode,
        price,
        weekly_price,
        monthly_price
    FROM melb_data_airbnb
      )
WHERE zipcode IS NOT NULL
GROUP BY zipcode
ORDER BY zipcode DESC
"""
with engine.connect() as con:
      rs = con.execute(query5)
      print(query5)
      for row in rs:
          print(row)
```

Ahora aplicamos esta consulta pero a la unión de las dos tablas

```python
query6 = """
SELECT * FROM melb_data
LEFT JOIN (
  SELECT zipcode ,  
    AVG(price) as price_mean, 
    AVG(weekly_price) as weekly_price_mean, 
    AVG(monthly_price) as monthly_price_mean
  FROM (
    SELECT 
        CAST(zipcode AS float) as zipcode,
        price,
        weekly_price,
        monthly_price
    FROM melb_data_airbnb
        )
  WHERE zipcode IS NOT NULL
  GROUP BY zipcode
  ORDER BY zipcode DESC
          )
      ON melb_data_airbnb.zipcode = melb_data.Postcode
"""
with engine.connect() as con:
      rs = con.execute(query6)
      print(query6)
      for row in rs:
          print(row)
```
